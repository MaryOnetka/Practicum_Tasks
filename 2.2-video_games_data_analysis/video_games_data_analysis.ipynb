### 2. Проверка ошибок в данных и их предобработка

 #### 2.1. Названия столбцов датафрейма

```python
# Выведем названия всех столбцов Датафрейма и проверим их стиль написания
df.columns
```

```python
# Приведем названия всех столбцов к стилю snake case
df.columns = df.columns.str.lower().str.replace(' ', '_')
df.columns
```

 #### 2.2. Типы данных

```python
# Уменьшим разрядность:
# преобразуем столбцы через pd.to_numeric для обработки ошибок и выбора разрядности
columns_name = ['na_sales', 'eu_sales', 'jp_sales', 'other_sales']
for column in columns_name:
    df[column] = pd.to_numeric(df[column], errors='coerce', downcast='float')
```

```python
# Столбец critic_score приводим к int8
# Пропускам присвоим значение-индикатор = -1.
df['critic_score'] = df['critic_score'].fillna(-1).astype('int8')
```

```python
# Преобразуем year_of_release тип к int16
# Пропускам присвоим значение-индикатор =-1
df['year_of_release'] = df['year_of_release'].fillna(-1).astype('int16')
```

```python
# Столбец user_score приводим к float
# Присвоим содержащимся в нем строковым значениям tbd пропуски:
df['user_score'] = pd.to_numeric(df['user_score'], errors='coerce').fillna(-1)
```

```python
# Выведем общую информацию о Датасете после проведения всех преобразований:
df.info()
```

Наш Датафрейм готов к дальнейшему анализу.

 #### 2.3. Наличие пропусков в данных

Посчитать количество пропусков в каждом столбце в абсолютном и относительном значениях.

```python
# Считаем абсолютное количество пропусков
df.isna().sum()
```

```python
# Считаем относительное количество пропусков
round(df.isna().mean()*100, 2)
```

```python
# В столбцах year_of_release, critic_score пропуски заменены на -1
# Посчитаем: сколько строк содержат -1

year_of_release_sum = len(df[df['year_of_release'] < 0])
critic_score_sum = len(df[df['critic_score'] < 0])
user_score_sum = len(df[df['user_score'] < 0])
year_of_release_per = len(df[df['year_of_release'] < 0]) / len(df) * 100
critic_score_per = len(df[df['critic_score'] < 0]) / len(df) * 100
user_score_per = len(df[df['user_score'] < 0]) / len(df) * 100

print(f" Количесвто пропусков в year_of_release: {year_of_release_sum}")
print(f" Количесвто пропусков в user_score: {user_score_sum}")
print(f" Количесвто пропусков в critic_score: {critic_score_sum}")
print(f" Соотношение пропусков в year_of_release: {round(year_of_release_per,2)}")
print(f" Соотношение пропусков в user_score: {round(user_score_per,2)}")
print(f" Соотношение пропусков в critic_score: {round(critic_score_per,2)}")
```

**Делаем выводы:**

Процент пропусков в столбцах `name`, `genre`, `eu_sales`, `jp_sales` мал, составляет менее 1% данных. Строки с этими пропусками можно удалить.

Проценты пропущенных значений в столбцах `user_score` - 54.66%, `rating` - 40.52%,  `critic_score` - 51.39% составляют почти половину данных в своих же столбцах. Такие данные просто удалить нельзя. 

Выявить точную закономерность не представляется возможным. Можно отметить только тот факт, что в большенстве случаев при отсутствии одной оценки, другие столбцы с оценками будут тоже пусты. 

Можно заменить такие пропуски усреднёнными значениями(рейтинг - мода по жанру игры, оценка - медиана по жанру игры, год предлагаю заменить на год с наибольшим количеством вышедших игр).

```python
# Удалим строки с пропусками в столбцах name, genre, eu_sales, jp_sales
df = df.dropna(subset=['name', 'genre', 'eu_sales', 'jp_sales']) 
```

```python
# Приведём названия жанров к нижнему регистру
df['genre'] = df['genre'].str.lower()
df['genre'].unique()
```

```python
# Заменим пропуски в столбце rating модой от жанра игры
df['rating'] = df['rating'].fillna(df.groupby('genre')['rating'].transform(lambda x: x.mode()[0]))
```

```python
# Заменим пропуски в столбце year_of_release модой
# Пропуск = -1
df['year_of_release'] = df['year_of_release'].replace(-1, int(df['year_of_release'].mode()[0]))
```

```python
# Контрольная проверка на наличие пропусков
df.isna().sum()
```

### 4. Категоризация данных

Необходимо категоризовать игры по оценкам пользователей и экспертов:
* **высокая оценка** — с оценкой от 8 до 10 и от 80 до 100, включая правые границы интервалов;
* **средняя оценка** — с оценкой от 3 до 8 и от 30 до 80, не включая правые границы интервалов;
* **низкая оценка** — с оценкой от 0 до 3 и от 0 до 30, не включая правые границы интервалов.

```python
# Создадим копию
new_period_df = new_period_df.copy()
```

```python
# Сделаем категоризацию на основе оценок пользователей
new_period_df['user_category'] = pd.cut(new_period_df['user_score'], bins=[-1,0, 3, 8, 10], labels=['без оценки','низкая оценка', 'средняя оценка', 'высокая оценка'], right=False)
```

```python
# Сделаем категоризацию на основе оценок критиков
new_period_df['critic_category'] = pd.cut(new_period_df['critic_score'], bins=[-1,0, 30, 80, 100], labels=['без оценки','низкая оценка', 'средняя оценка', 'высокая оценка'], right=False)
```

```python
# Выведем таблицу после категоризаций
new_period_df
```

Игры разбиты на категории на основе оценок пользователей и критиков. Можно сгруппировать полученные данные для подсчёта количества игр.

```python
#Количество игр по категориям на основе оценок пользователей
new_period_df.groupby('user_category', observed=False)['name'].count()
```

```python
#Количество игр по категориям на основе оценок критиков
new_period_df.groupby('critic_category', observed=False)['name'].count()
```

В разрезе двух категоризаций большинство игр находится в пункте **средняя оценка**.
Меньше всего игр находится в категории **низкая оценка**(0.91% от общего количества оценок пользователя и 0.45% - от оценок критиков).
**Высокую оценку** игры от критиков получали меньше раз, чем от пользоватей.
**Без оценки** в обеих категоризациях составляют довольно большое число, скорее всего из-за неактуальности платформ и площадок сбора информации этих игр, возможно и региональные распределения сыграли свою роль.

Выделим **ТОП-7** платформ по количеству игр, выпущенных за период с 2000 года по 2013 год.

```python
new_period_df.groupby('platform')['name'].count().sort_values(ascending=False).head(7)
```

Первых два фаворита:
* PS2 - Sony Playstation 2 (2160 игр);
* Nintendo DS (2147 игр).
